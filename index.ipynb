## Question 2

Explain the relationship between **Primary Keys** and **Foreign Keys**.

Type your answer below this line:
_______________________________________________________________________________________________________________________________

A primary key is a unique identifier for a table ie id number while foreign key is a primary key from a different table.
The purpose of a primary key is to uniquely identify each row in a table. This is important because it allows us to easily find and update specific rows in the table. The purpose of a foreign key is to establish a relationship between two tables. This is important because it allows us to relate data from different tables.

.	One-to-one: This relationship occurs when one row in a table can only be associated with one row in another table. 
.	One-to-many: This relationship occurs when one row in a table can be associated with many rows in another table. For example, a customer can order many products.
.	Many-to-many: This relationship occurs when many rows in one table can be associated with many rows in another table. For example, a student can take many courses, and a course can be taken by many students.
 




## Question 4

Explain the various types of JOINs possible with SQL. 

Type your answer below this line:
_______________________________________________________________________________________________________________________________

INNER JOIN;
The INNER JOIN returns only the matching rows from both tables based on the specified join condition.
It selects rows where the join condition is satisfied in both tables.

LEFT JOIN;
The LEFT JOIN returns all the rows from the left (or first) table and the matching rows from the right (or second) table based on the join condition.
If there are no matches in the right table, NULL values are returned for the right table's columns.

RIGHT JOIN;
The RIGHT JOIN returns all the rows from the right (or second) table and the matching rows from the left (or first) table based on the join condition.
If there are no matches in the left table, NULL values are returned for the left table's columns.



## Question 5

Explain the relationship between Aggregate functions and GROUP BY statements.

Type your answer below this line:
_______________________________________________________________________________________________________________________________

Aggregate functions are functions that operate on a set of values and return a single value. Some common aggregate functions include COUNT(), SUM(), AVG(), and MAX().
GROUP BY statements are used to group rows together based on a common value. 

The relationship between aggregate functions and GROUP BY statements is that aggregate functions can only be used in conjunction with GROUP BY statements. This is because aggregate functions operate on groups of rows, and GROUP BY statements are used to group rows together.


## Question 6

What role do Associative Entities play (JOIN Tables) in many-to-many JOINs?


Type your answer below this line:
_______________________________________________________________________________________________________________________________

The associative entity, or join table, acts as an intermediary to resolve many-to-many relationships by establishing the connections between the two main entities. It plays a vital role in maintaining data integrity, storing additional attributes specific to the relationship, and facilitating querying and manipulation of data in many-to-many joins.
